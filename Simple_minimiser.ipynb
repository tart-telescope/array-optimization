```python
import torch
import matplotlib.pyplot as plt

# Define the function y = (x-2)^2 + 4
def loss_function(x):
    return ((x - 2) ** 2) + 4

# Initialize x as a tensor with requires_grad=True to track computations
x = torch.tensor([0.0], requires_grad=True)

# Set the learning rate
learning_rate = 0.3

# Define the number of optimization steps
steps = 1000

# Use the Adam optimizer from PyTorch
optimizer = torch.optim.SGD([x], lr=learning_rate)

loss_history = []
# Optimization loop
for step in range(steps):
    # Zero the gradients from the previous step
    optimizer.zero_grad()

    # Compute the function value at the current x
    y = loss_function(x)

    # Perform backpropagation to compute the gradient of y with respect to x
    y.backward()

    # Update x using the computed gradient
    optimizer.step()

    loss_history.append(loss_function(x).item())

    # Print the value of x and y every 100 steps
    if step % 100 == 0:
        print(f'Step {step}: x = {x.item()}, y = {y.item()}')

# Print the final optimized value of x and the minimum value of y
print(f'Optimized x: {x.item()}')
print(f'Minimum y: {loss_function(x).item()}')

plt.plot(loss_history, 'x')
plt.title("Loss vs iteration")
plt.ylabel('Loss')
plt.xlabel('Iternation')
plt.grid(True)
plt.yscale('log')
plt.xscale('log')
```
